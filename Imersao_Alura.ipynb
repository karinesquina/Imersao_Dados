## **Aula 04**

Imagina que você precisa determinar o valor do seu imóvel para vendê-lo. Para isso você analisa algumas variáveis, tipo: quantidade de quartos, m², localização, etc. A partir da análise dessas variáveis, você vai prever o valor que você vai cobrar pela venda do seu apartamento. É exatamente o que a ML faz (regressão).  Mas levando em consideração o caso acima, como você chegou na conclusão desses valores? Por exemplo, como você sabe que quanto mais quartos, mais valorizado seu imóvel? Você, tem uma base de conhecimento, pelas pesquisas que você fez durante sua vida. Ou seja, você foi treinado para saber que, quanto mais quartos no seu imovel, mais valorizado ele é. Esse treino foi válido na hora que você foi testado para preçificar seu imóvel. É por isso que quando vamos trabalhar com ML, separamos uma amostra para treino, e uma amostra para teste.

```
#Limpando os dados de Nan da minha amostra. 
#esse passo sera importante na hora de aplicar o fit.
dados_sem_zerados = dados_sem_zerados[provas].dropna() #eu tenho que informar que eu quero remover apenas os Nan das provas, para ele nao remover colunas ou linhas que outros dados.
```

```
#prever a nota da prova MT. Separo as provas de entrada (que são os dados para eu comparar) e as provas de saida (o dado que eu quero prever)
provas_entrada = ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']
prova_saida = 'NU_NOTA_MT'

notas_entrada = dados_sem_zerados[provas_entrada]
notas_saida = dados_sem_zerados[prova_saida]

notas_entrada
```

```
x = notas_entrada
y = notas_saida
```

```
#fazer separação de base de treino e base de teste.Test_size = tamanho da amostra. Random_state = pegar sempre a mesma amostra.
from sklearn.model_selection import train_test_split
seed = 1234 #garantir que a amostra virá sempre a mesma.
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.25, random_state = seed)
x_treino.head()
```

```
#verificar o tamanho da amostra. (numero de linhas x numero de colunas)
x_treino.shape
```

```
#contar a quantidade de linhas. Nesse caso, o y_treino tem que ter o mesmo valor do x_treino.
len(y_treino)
```

```
#Criar um modelo de regressão, ou seja, criar um modelo que vai prever minhas notas de MT. 
#modelo escolhido = regressão linear.
#SVM = suporter vector machine (treinar o modelo)
from sklearn.svm import LinearSVR

modelo = LinearSVR(random_state = seed) #precisa do random_state pois ele também gera numeros randomicos.
modelo.fit(x_treino, y_treino) #pega meus dados de modelo, e se ajusta a ele. Ex. Quando vc entra em um carro, e vc ajusta o banco, espelhos, retrovisor..etc. para começar a dirigir.
```

```
#calcular a predição (afirmar o que vai acontecer no futuro) através das notas que eu já possuo.
predicao_matematica = modelo.predict(x_teste)
predicao_matematica
#cada valor retornado significa: baseado na sua nota X, sua nota de MT seria...ele faz isso com todas as notas q eu estou comparando.
```

```
#comparando minha base de treino com o meu teste. 
#Lembrando que meu X são os dados q eu estou usando para treinar, e o meu y é o meu dado real (a nota de MT que já existe).
y_teste[:6]
```

Analisando os resultados do predict vejo que, das 5 predições, as predicoes 2 e 3 foram as que chegaram mais proximo da minha realidade.


```
sns.scatterplot(x = predicao_matematica, y = y_teste)
```

```
#Calcular o erro de predição. Meu eixo X vai ser a nota real e no eixo y vai ser o quanto eu errei na minha predicao.
sns.scatterplot(x = y_teste, y = y_teste - predicao_matematica)
```

Meu erro foi equilibrado, ou seja, eu chutei mais alto e mais baixo do que o valor real. A análise me mostra que, quanto menor a nota da pessoa, meu modelo abaixou ainda mais a nota de matematica dela. E quanto maior a nota dela, mais alto foi a predição da nota de matemática.

```
#outra maneira de analisar. Subtrair a média do meu x_teste do meu y_teste (se usar a predição).
#axis = coluna
sns.scatterplot(x = y_teste, y = x_teste.mean(axis=1))
```

Nesse caso eu estou chutando que se a pessoa acertou 10, 9 e 10.. a media dela vai ser 9.5. 

```
#construir a métrica para avaliar o modelo. 
resultados = pd.DataFrame() #criando um dataframe. Dataframe recebe varias listas.
resultados['Real'] = y_teste #criando uma coluna chamada Real para atribuir o valor de y_teste
resultados['Predicao'] = predicao_matematica #criando outra coluna chamada Predicao.
resultados['Diferenca'] = resultados['Real'] - resultados['Predicao'] #calculando a diferença e criando outra coluna
resultados['Quadrado_diferenca'] = resultados['Diferenca']**2 # para tirar os valores negativos e melhorar a perfomance de análise.
resultados
```

Analisando a situação acima, quanto mais proximo do 0, melhor a minha análise. Pois numeros muito acima ou abaixo do 0, destoam da realidade. Em relação ao quadrado_diferença, eu nao posso somar os valores negativos com os positivos (para fazer a média, por exemplo). Vamos dizer que eu tenha uma amostra com 3 pessoas, e eu fiz as seguintes previsões de doença:
pessoa 1 = -100;
pessoa 2 = 0;
pessoa 3 = 100;
Se eu for tirar uma média dessa minha previsão, eu vou dizer que eu fiquei em 0, ou seja, que meu modelo foi muito proximo da realidade. Mas na verdade, nao, pois o que aconteceu foi que 2 erros foram anulados. Entao, para que eu tenha certeza que meu modelo esta proximo do 0, todos os meus valores precisam ser positivos, pois assim, um erro vai somar com outro. 


```
#a métrica para calcular o modelo, é calculada com a média. 
resultados['Quadrado_diferenca'].mean()
```

```
#para eu saber se esse valor é muito alto, preciso fazer a raiz dele. 
resultados['Quadrado_diferenca'].mean()**(1/2)
```

Os valores da Arvore de decisão acabam sendo mais seguros, pois eles variam muito pouco quando rodados novamente. Mas analisando todo o contexto, verificamos que quando fazemos o split para dividir os dados de treino e teste, também temos esse fator de aleatoriedade. Para que possamos verificar com mais segurança esses resultados, vamos treinar nosso modelo vária vezes, dividindo ele em várias repartições.

```
#dividindo o modelo de treino em várias repartições, e fazendo a média deles. 
from sklearn.model_selection import cross_validate
modelo_arvore = DecisionTreeRegressor(max_depth=2)
cross_validate(modelo_arvore, x, y)
```

Analisando os resultados, verificamos que ele retornou 3 tipos de resultados. O fit_time e o score_time, são os tempos que levaram para rodar o modelo. O test_score é a amostra da divisão do treino, por padrão ele divide em 5 pontos.

```
#aplicando o contador de erro. Preciso usar o neg na frente.Quanto mais proximo de 0 o valor, melhor.
cross_validate(modelo_arvore, x, y, scoring='neg_mean_squared_error')
```

Quando vc aplica o mean_squared_error no scoring o valor que retorna é sempre negativo, pq quando vc fala de scoring, vc esta dizendo que quanto maior, melhor. Então para que não houvesse confusão na hora de analisar, os desenvolvedores optaram por deixar os numeros negativos, assim sabemos que quanto maior o numero (ou no caso de numeros negativos, quanto mais proximo do zero), melhor a avaliação do modelo. Lembrando que o mean_squared_error nao é uma métrica de score, ele calcula o erro, mas nesse caso eu passei ele como parametro de scoring, por isso essa avaliação negativa.

```
#apresentando o resultado apenas do que eu quero saber, que é o test_score. Posso transformar o valor em positivo multiplicando por *-1
resultado = cross_validate(modelo_arvore, x, y, scoring='neg_mean_squared_error')
resultado['test_score']*-1
```

```
#aplicando a media
media = (resultado['test_score']*-1).mean()
media
```

```
#analisando os resultados com mais divisoes. Aplicando o cv.
resultado = cross_validate(modelo_arvore, x, y, cv= 10, scoring='neg_mean_squared_error')
resultado['test_score']*-1
```

```
#calculando o desvio padrão e os limites de 95% da amostra.
resultado = cross_validate(modelo_arvore, x, y, cv= 10, scoring='neg_mean_squared_error')
media = (resultado['test_score']*-1).mean()
desvio_padrao = (resultado['test_score']*-1).std()
lim_inferior = media - (2*desvio_padrao)
lim_superior = media + (2*desvio_padrao)

print(f'Intervalo de confiança {lim_inferior} - {lim_superior}')
```

```
#definir uma função com o calculo acima.
def calcula_mse(resultado):
    resultado = cross_validate(modelo_arvore, x, y, cv= 10, scoring='neg_mean_squared_error')
    media = (resultado['test_score']*-1).mean()
    desvio_padrao = (resultado['test_score']*-1).std()
    lim_inferior = media - (2*desvio_padrao)
    lim_superior = media + (2*desvio_padrao)
    print(f'Intervalo de confiança {lim_inferior} - {lim_superior}')
```

Quando eu rodo essa função eu sempre vou ter os mesmos resultados. Para que eu possa verificar a eficácia do meu modelo, eu posso aplicar uma função para embaralhar os dados.

```
#embaralhar os dados e trazer resultados diferentes, para comparação.
from sklearn.model_selection import KFold

partes_embaralhadas = KFold(n_splits= 10, shuffle= True) #n_split é a quantidade de partes que eu quero o shuffle embaralha os valores.
modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultado = cross_validate(modelo_arvore, x, y, cv= partes_embaralhadas, scoring='neg_mean_squared_error')
calcula_mse(resultado)
```

```
#aplicar um padrao para essa variação, usando o seed.
from sklearn.model_selection import KFold
import numpy as np

seed= 1234
np.random.seed(seed)

partes_embaralhadas = KFold(n_splits= 10, shuffle= True) #n_split é a quantidade de partes que eu quero o shuffle embaralha os valores.
modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultado = cross_validate(modelo_arvore, x, y, cv= partes_embaralhadas, scoring='neg_mean_squared_error')
calcula_mse(resultado)
```

```
#colocando meu cálculo dentro de uma função. Definindo o nível que vou querer de divisão. E apresentando a média.
def regressor_arvore(nivel):
    seed= 1234
    np.random.seed(seed)

    partes_embaralhadas = KFold(n_splits= 10, shuffle= True) #n_split é a quantidade de partes que eu quero o shuffle embaralha os valores.
    modelo_arvore = DecisionTreeRegressor(max_depth=nivel)
    resultado = cross_validate(modelo_arvore, x, y, cv= partes_embaralhadas, scoring='neg_mean_squared_error')
    print(f"Teste = {(resultado['test_score']*-1).mean()}")

regressor_arvore(4)
```

```
#avaliar o nivel de profundidade de 1 a 20.
for i in range(1, 21):
    regressor_arvore(i)
```

Analisando o resultado do for, eu percebo que em um determinado momento meus resultados estão aumentando, e isso nao é bom, pois preciso de um valor proximo de zero. Então eu vou comparar esses dados com os dados de treino.

```
#comparando os dados de treino e teste. Pego a mesma funcao acima e acrescento um return_train_score.
def regressor_arvore(nivel):
    seed= 1234
    np.random.seed(seed)

    partes_embaralhadas = KFold(n_splits= 10, shuffle= True) #n_split é a quantidade de partes que eu quero o shuffle embaralha os valores.
    modelo_arvore = DecisionTreeRegressor(max_depth=nivel)
    resultado = cross_validate(modelo_arvore, x, y, cv= partes_embaralhadas, scoring='neg_mean_squared_error', return_train_score=True)
    print(f"Treino = {(resultado['train_score']*-1).mean()} | Teste = {(resultado['test_score']*-1).mean()}")

regressor_arvore(4)
```

```
#avaliar o nivel de profundidade de 1 a 20.
for i in range(1, 21):
    regressor_arvore(i)
```

Na avaliação do for, eu identifiquei que meu treino vai diminuindo e meu teste, diminui ate uma parte, e começa a aumentar novamente. Isso significa que ele se viciou na base de treino, e por isso ele deixou de ser eficiente no teste, pois ele decorou os dados de treino. Isso é um dos parametros do Overfit.
